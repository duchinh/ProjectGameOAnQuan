### 2.4. Giải thuật MINIMAX

Sử dụng giải thuật minimax có dùng cắt tỉa anpha và beta để giảm bớt trường hợp tính toán thừa

- Minimax (hay còn gọi là minmax) là một phương pháp trong lý thuyết quyết định
  có mục đích là tối thiểu hóa (minimiza) tổn thất vốn được dự tính có thể là “tối đa”
  (maximize). Có thể hiểu ngược lại là nó nhằm tối đa hóa lợi ích vốn được dự tính là tối
  thiểu (maximin).

* Hai đối thủ trong trò chơi có tên là MAX và MIN.

- Max: biểu diễn cho mục đích của đối thủ này là làm lớn tối đa lợi thế của
  mình.
- Min: biểu diễn cho mục đích của đối thủ này là làm nhỏ tối đa lợi thế của
  đối phương.
  Trên cây tìm kiếm sẽ phân lớp thành các lớp Max và Min.

* Giải thuật cắt tỉa Alpha-Beta giảm được nhiều nhánh tìm kiếm vô ích, giúp tăng tốc
  độ tìm kiếm, giảm chi phí thực thi và tăng độ sâu tìm kiếm, từ đó giá trị của hàm tiện
  ích

3.2. Áp dụng

Áp dụng chiến lược Minimax, Hai đối thủ trong một trò chơi được gọi là MIN và
MAX. MAX đại diện cho đối thủ quyết giành thắng lợi hay cố gắng tối đa hóa ưu thế
của mình. Ngược lại MIN là đối thủ cố gắng tối thiểu hóa điểm số của MAX. Ở đây,
dựa trên ý tưởng là tối đa và tối thiểu hiệu số giữa số quân ăn được của MAX và số
quân ăn được của MIN.

Chiến lược Minimax thể hiện qua quy tắc định giá trị cho các nút trên cây trò chơi
như sau:

- Nếu nút là nút lá gán cho nút đó một giá trị để phản ánh trạng thái thắng thua hay hòa
  của các đấu thủ.
- Sử dụng giá trị của các nút lá để xác định giá trị của các nút ở các mức trên trong cây
  trò chơi theo quy tắc:

* Nút thuộc lớp MAX thì gán cho nó giá trị lớn nhất của các nút con của nút đó.
* Nút thuộc lớp MIN thì gán cho nó giá trị nhỏ nhất của các nút con của nút đó


Ta xây dựng giải thuật Minimax như sau:


````python
public int minimax(INode node, int depth, boolean maximizingPlayer) {
		if (depth == 0 || (node.getAI().isOver() && checkScore(node) > 0)) {
			return heuristic(node);
		}
		if (maximizingPlayer) {
			int Mx = Integer.MIN_VALUE;
			for (INode n : node.getNeighbours(maximizingPlayer)) {
				Mx = Math.max(Mx, minimax(n, depth - 1, false));
			}
			return Mx;
		} else {
			int Mn = Integer.MAX_VALUE;
			for (INode n : node.getNeighbours(maximizingPlayer)) {
				Mn = Math.min(Mn, minimax(n, depth - 1, true));
			}
			return Mn;
		}

	}
````

Ngoài ra, hàm Heuristic để đánh giá độ tốt của một trạng thái trên bàn cờ, của
một người chơi nào đó (MIN hay MAX, 1 hay -1). Đây là hàm quyết định, chi phối hầu
hết thời gian và chỉ làm việc trên các nút lá, vì vậy việc thăm các nút không phải các
nút lá có thể bỏ qua.


````python
public int heuristic(INode node) {
    return node.getAI().getListPlayer().get(0).getScore() - node.getAI().getListPlayer().get(1).getScore();
}
````

### 3.3. Giải thuật alpha, beta:

Giải thuật alpha, beta là một cái tiến của giải thuật Minimax nhằm tỉa nhánh của
cây trò chơi, làm giảm số lượng nút phải sinh và lượng giá, do đó có thể tăng độ sâu của
cây tìm kiếm.
Ý tưởng của tìm kiếm Alpha-beta rất đơn giản: Có hai giá trị, gọi là alpha và beta
được tạo ra trong quá trình tìm kiếm:

- Giá trị alpha liên quan với các nút MAX và có khuynh hướng không bao giờ
  giảm.
- Ngược lại giá trị beta liên quan đến các nút MIN và có khuynh hướng không
  bao giờ tăng.

Từ đó ta có luật cắt tỉa dựa trên các giá trị alpha và beta:

- Quá trình tìm kiếm có thể kết thúc bên dưới một nút MIN nào có giá trị beta
  nhỏ hơn hoặc bằng giá trị alpha của một nút cha MAX bất kỳ của nó.
- Quá trình tìm kiếm có thể kết thúc bên dưới một nút MAX nào có giá trị alpha
  lớn hơn hoặc bằng giá trị beta của một nút cha MIN bất kỳ của nó.
  Từ ý tưởng và luật trên ta sẽ xây dựng hàm cải tiến từ hàm Minimax phía trên có
  cắt tỉa alpha, beta.


````python
public int alphabeta(INode node, int depth, int a, int b, boolean maximizingPlayer) {
    if (depth == 0 || (node.getAI().isOver() && checkScore(node) > 0)) {
        return heuristic(node);
    }
    if (maximizingPlayer) {
        for (INode n : node.getNeighbours(maximizingPlayer)) {
            a = Math.max(a, alphabeta(n, depth - 1, a, b, false));
            if (a >= b)
                break;
        }
        return a;
    } else {
        for (INode n : node.getNeighbours(maximizingPlayer)) {
            b = Math.min(b, alphabeta(n, depth - 1, a, b, true));
            if (a >= b)
                break;
        }
        return b;
    }
}
````

Thuật giải Alpha-beta nói chung giúp ta tiết kiệm nhiều thời gian so với Minimax
mà vẫn đảm bảo kết quả tìm kiếm chính xác. Tuy nhiên lượng tiết kiếm này không ổn
định – phụ thuộc vào số nút mà nó cắt bỏ. Trong trường hợp xấu nhất, giải thuật không
cắt được một nhánh nào và phải xét số nút lá bằng với giải thuật Minimax. Vì thế ta cần
đẩy mạnh việc cắt bỏ nhờ đẩy nhanh sự thu hẹp của hai giá trị alpha và beta. Khoảng
cách này được thu hẹp một bước khi gặp một giá trị mới tốt hơn giá trị cũ. Như vậy,
nếu gặp giá trị tốt nhất càng sớm, thì số nhánh được cắt bỏ càng nhiều.

### 3.7.  CÁC VẤN ĐỀ VÀ THUẬT GIẢI

#### Xây dựng hàm Minimax có cắt tỉa (Minimax Algorithm with Alpha-Beta Pruning)
### Giải thích hàm `minimax`

Hàm `minimax` là một thuật toán được sử dụng trong trí tuệ nhân tạo để đưa ra quyết định tốt nhất trong các trò chơi hai người chơi, chẳng hạn như cờ vua, cờ vây, hoặc trong trường hợp này là một trò chơi với các ô và đá. Thuật toán này giúp AI đánh giá và chọn nước đi tối ưu bằng cách mô phỏng các nước đi tiếp theo và đánh giá kết quả của chúng.

#### Cấu trúc và các thành phần của hàm `minimax`

```java
public int minimax(INode node, int depth, boolean maximizingPlayer) {
    // Điều kiện dừng của thuật toán: độ sâu bằng 0 hoặc trò chơi đã kết thúc và điểm số lớn hơn 0
    if (depth == 0 || (node.getAI().isOver() && checkScore(node) > 0)) {
        return heuristic(node); // Trả về giá trị heuristic của node
    }
    
    // Nếu đang ở lượt của người chơi tối đa hóa
    if (maximizingPlayer) {
        int Mx = Integer.MIN_VALUE;
        // Duyệt qua các node con của node hiện tại
        for (INode n : node.getNeighbours(maximizingPlayer)) {
            // Gọi đệ quy hàm minimax cho các node con và cập nhật giá trị tối đa
            Mx = Math.max(Mx, minimax(n, depth - 1, false));
        }
        return Mx; // Trả về giá trị tối đa
    } else {
        // Nếu đang ở lượt của người chơi tối thiểu hóa
        int Mn = Integer.MAX_VALUE;
        // Duyệt qua các node con của node hiện tại
        for (INode n : node.getNeighbours(maximizingPlayer)) {
            // Gọi đệ quy hàm minimax cho các node con và cập nhật giá trị tối thiểu
            Mn = Math.min(Mn, minimax(n, depth - 1, true));
        }
        return Mn; // Trả về giá trị tối thiểu
    }
}
```


1. **Điều kiện dừng:**
   ```java
   if (depth == 0 || (node.getAI().isOver() && checkScore(node) > 0)) {
       return heuristic(node);
   }
   ```
   - Điều kiện này kiểm tra xem độ sâu (depth) đã bằng 0 hoặc trò chơi đã kết thúc và điểm số của node hiện tại lớn hơn 0.
   - Nếu một trong hai điều kiện này đúng, hàm sẽ trả về giá trị heuristic của node hiện tại. Giá trị heuristic là một đánh giá của trạng thái hiện tại của trò chơi, cho biết mức độ tốt xấu của trạng thái này.

2. **Maximizing Player:**
   ```java
   if (maximizingPlayer) {
       int Mx = Integer.MIN_VALUE;
       for (INode n : node.getNeighbours(maximizingPlayer)) {
           Mx = Math.max(Mx, minimax(n, depth - 1, false));
       }
       return Mx;
   }
   ```
   - Nếu `maximizingPlayer` là `true`, AI sẽ cố gắng tối đa hóa điểm số.
   - Biến `Mx` được khởi tạo với giá trị nhỏ nhất có thể (`Integer.MIN_VALUE`).
   - AI sẽ duyệt qua tất cả các node con của node hiện tại (`node.getNeighbours(maximizingPlayer)`), gọi đệ quy hàm `minimax` với `maximizingPlayer` là `false` (chuyển sang lượt của người chơi đối thủ), và cập nhật giá trị `Mx` với giá trị lớn nhất giữa `Mx` hiện tại và giá trị trả về của hàm `minimax` gọi đệ quy.
   - Trả về giá trị `Mx`, đại diện cho nước đi tốt nhất cho người chơi tối đa hóa tại node hiện tại.

3. **Minimizing Player:**
   ```java
   else {
       int Mn = Integer.MAX_VALUE;
       for (INode n : node.getNeighbours(maximizingPlayer)) {
           Mn = Math.min(Mn, minimax(n, depth - 1, true));
       }
       return Mn;
   }
   ```
   - Nếu `maximizingPlayer` là `false`, AI sẽ cố gắng tối thiểu hóa điểm số.
   - Biến `Mn` được khởi tạo với giá trị lớn nhất có thể (`Integer.MAX_VALUE`).
   - AI sẽ duyệt qua tất cả các node con của node hiện tại (`node.getNeighbours(maximizingPlayer)`), gọi đệ quy hàm `minimax` với `maximizingPlayer` là `true` (chuyển sang lượt của người chơi tối đa hóa), và cập nhật giá trị `Mn` với giá trị nhỏ nhất giữa `Mn` hiện tại và giá trị trả về của hàm `minimax` gọi đệ quy.
   - Trả về giá trị `Mn`, đại diện cho nước đi tốt nhất cho người chơi tối thiểu hóa tại node hiện tại.

#### Tóm tắt:
- Hàm `minimax` giúp AI đưa ra quyết định tối ưu bằng cách mô phỏng các nước đi có thể và đánh giá kết quả của chúng.
- AI sẽ cố gắng tối đa hóa điểm số của mình và tối thiểu hóa điểm số của đối thủ bằng cách duyệt qua các node con và gọi đệ quy hàm `minimax`.
- Điều kiện dừng đảm bảo hàm không chạy vô hạn và trả về giá trị đánh giá của trạng thái hiện tại khi đạt độ sâu nhất định hoặc khi trò chơi kết thúc.

### Xây dựng hàm Alpha-Beta 

### Giải thích hàm `alphabeta`

Hàm `alphabeta` là một phiên bản tối ưu hóa của thuật toán Minimax được gọi là Alpha-Beta Pruning. Mục tiêu của Alpha-Beta Pruning là giảm số lượng node cần phải đánh giá trong cây tìm kiếm của thuật toán Minimax, bằng cách cắt tỉa (pruning) những nhánh không cần thiết. Điều này giúp giảm thời gian tính toán và làm cho thuật toán hiệu quả hơn.

#### Cấu trúc và các thành phần của hàm `alphabeta`

```java
public int alphabeta(INode node, int depth, int a, int b, boolean maximizingPlayer) {
    // Điều kiện dừng của thuật toán: độ sâu bằng 0 hoặc trò chơi đã kết thúc và điểm số lớn hơn 0
    if (depth == 0 || (node.getAI().isOver() && checkScore(node) > 0)) {
        return heuristic(node); // Trả về giá trị heuristic của node
    }
    
    if (maximizingPlayer) {
        // Duyệt qua các node con của node hiện tại
        for (INode n : node.getNeighbours(maximizingPlayer)) {
            // Gọi đệ quy hàm alphabeta cho các node con và cập nhật giá trị alpha
            a = Math.max(a, alphabeta(n, depth - 1, a, b, false));
            if (a >= b) // Cắt tỉa các nhánh không cần thiết
                break;
        }
        return a; // Trả về giá trị alpha
    } else {
        // Duyệt qua các node con của node hiện tại
        for (INode n : node.getNeighbours(maximizingPlayer)) {
            // Gọi đệ quy hàm alphabeta cho các node con và cập nhật giá trị beta
            b = Math.min(b, alphabeta(n, depth - 1, a, b, true));
            if (a >= b) // Cắt tỉa các nhánh không cần thiết
                break;
        }
        return b; // Trả về giá trị beta
    }
}
```

#### Chi tiết các thành phần:

1. **Điều kiện dừng:**
   ```java
   if (depth == 0 || (node.getAI().isOver() && checkScore(node) > 0)) {
       return heuristic(node);
   }
   ```
   - Điều kiện này kiểm tra xem độ sâu (depth) đã bằng 0 hoặc trò chơi đã kết thúc và điểm số của node hiện tại lớn hơn 0.
   - Nếu một trong hai điều kiện này đúng, hàm sẽ trả về giá trị heuristic của node hiện tại. Giá trị heuristic là một đánh giá của trạng thái hiện tại của trò chơi, cho biết mức độ tốt xấu của trạng thái này.

2. **Maximizing Player:**
   ```java
   if (maximizingPlayer) {
       for (INode n : node.getNeighbours(maximizingPlayer)) {
           a = Math.max(a, alphabeta(n, depth - 1, a, b, false));
           if (a >= b)
               break;
       }
       return a;
   }
   ```
   - Nếu `maximizingPlayer` là `true`, AI sẽ cố gắng tối đa hóa điểm số.
   - Biến `a` (alpha) được cập nhật với giá trị lớn nhất giữa `a` hiện tại và giá trị trả về của hàm `alphabeta` gọi đệ quy với `maximizingPlayer` là `false` (chuyển sang lượt của người chơi đối thủ).
   - Nếu giá trị `a` lớn hơn hoặc bằng `b` (beta), các nhánh còn lại sẽ bị cắt tỉa (không cần đánh giá nữa).
   - Trả về giá trị `a`, đại diện cho giá trị tốt nhất mà người chơi tối đa hóa có thể đạt được tại node hiện tại.

3. **Minimizing Player:**
   ```java
   else {
       for (INode n : node.getNeighbours(maximizingPlayer)) {
           b = Math.min(b, alphabeta(n, depth - 1, a, b, true));
           if (a >= b)
               break;
       }
       return b;
   }
   ```
   - Nếu `maximizingPlayer` là `false`, AI sẽ cố gắng tối thiểu hóa điểm số.
   - Biến `b` (beta) được cập nhật với giá trị nhỏ nhất giữa `b` hiện tại và giá trị trả về của hàm `alphabeta` gọi đệ quy với `maximizingPlayer` là `true` (chuyển sang lượt của người chơi tối đa hóa).
   - Nếu giá trị `a` lớn hơn hoặc bằng `b`, các nhánh còn lại sẽ bị cắt tỉa (không cần đánh giá nữa).
   - Trả về giá trị `b`, đại diện cho giá trị tốt nhất mà người chơi tối thiểu hóa có thể đạt được tại node hiện tại.

#### Tóm tắt:
- Hàm `alphabeta` giúp AI đưa ra quyết định tối ưu bằng cách mô phỏng các nước đi có thể và đánh giá kết quả của chúng, nhưng hiệu quả hơn nhờ vào việc cắt tỉa các nhánh không cần thiết.
- AI sẽ cố gắng tối đa hóa điểm số của mình và tối thiểu hóa điểm số của đối thủ bằng cách duyệt qua các node con và gọi đệ quy hàm `alphabeta`.
- Biến `alpha` (a) và `beta` (b) giúp kiểm soát quá trình cắt tỉa, giảm số lượng node cần phải đánh giá, từ đó cải thiện hiệu suất của thuật toán.

#### 3.8. Các hàm xử lí cơ bản trong game (Basic Game Handling Functions)
- **isValid(int index, boolean turn)**
  - Kiểm tra tính hợp lệ của một nước đi dựa trên vị trí và lượt chơi.
- **move(int index, boolean direction, boolean turn)**
  - Thực hiện di chuyển một ô trên bàn cờ.
- **checkBox(boolean turn)**
  - Kiểm tra tình trạng các ô dựa trên lượt chơi.
- **addChess(boolean turn)**
  - Thêm đá vào phần sân của người chơi.
- **isOver()**
  - Kiểm tra xem trò chơi có kết thúc chưa.
- **gameOver()**
  - Xử lý khi trò chơi kết thúc, cập nhật điểm số và hiển thị kết thúc trò chơi.
